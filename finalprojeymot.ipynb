```python
y = "HeartDisease"  # Hedef değişken
x = [col for col in data.columns if col != y]  # Girdi değişkenleri
```

```python
from h2o.automl import H2OAutoML

# AutoML başlat (örnek: 5 dakikalık süre)
aml = H2OAutoML(max_runtime_secs=300, seed=1)
aml.train(x=x, y=y, training_frame=data)
```

```python
data['HeartDisease'] = data['HeartDisease'].asfactor()
```

```python
aml = H2OAutoML(max_runtime_secs=600, seed=1)
aml.train(x=x, y=y, training_frame=data)
```

```python
predictions = aml.leader.predict(data)
predictions.head()
```
